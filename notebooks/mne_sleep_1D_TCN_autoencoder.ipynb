```
import matplotlib.pyplot as plt
import numpy as np
import csv
import os
import mne
from mne.datasets.sleep_physionet.age import fetch_data

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import classification_report
from tensorflow.keras.layers import Input, BatchNormalization, UpSampling1D, MaxPooling1D, Flatten, Dense, LeakyReLU, Dropout, Conv1D, Reshape
from tensorflow.keras.regularizers import l2
from tensorflow.keras.models import Model, load_model, save_model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import LearningRateScheduler
from tcn import compiled_tcn, TCN
from tensorflow.keras import backend as K
```

## DATASET PREPROCESSING

```
def build_dataset(start, end):
  dataset = {'x': [], 'y':[]}
  people = [i for i in range(start, end)]
  sf = 100
  annotation_desc_2_event_id = {'Sleep stage W': 1,
                                'Sleep stage 1': 2,
                                'Sleep stage 2': 3,
                                'Sleep stage 3': 4,
                                'Sleep stage 4': 4,
                                'Sleep stage R': 5}
  event_id = {'Sleep stage W': 1,
              'Sleep stage 1': 2,
              'Sleep stage 2': 3,
              'Sleep stage 3/4': 4,
              'Sleep stage R': 5}
  
  tmax = 30. - 1. / sf
  
  files = fetch_data(subjects=people, recording=[1])
  
  count = 0
  for file in files:
      raw = mne.io.read_raw_edf(file[0])
      annot = mne.read_annotations(file[1])
      raw.set_annotations(annot, emit_warning=False)
      events, _ = mne.events_from_annotations(
          raw, event_id=annotation_desc_2_event_id, chunk_duration=30.)
      epochs = mne.Epochs(raw=raw, events=events,
                                event_id=event_id, tmin=0., tmax=tmax, baseline=None)
      
      if count == 0:
          X = (epochs.get_data(picks="eeg"))[:,0,:]
          Y = epochs.events[:, 2]
          count += 1
      else:
          x = (epochs.get_data(picks="eeg"))[:,0,:]
          y = epochs.events[:, 2]
          X = np.concatenate((X, x), axis=0)
          Y = np.concatenate((Y, y), axis=0)
      
  dataset['x'] = X
  dataset['y'] = Y
  return dataset
```

```
def normalize_vector(vector):
  normalized_vector = (vector - vector.min()) / (vector.max() - vector.min())
  return normalized_vector
```

```
dataset = build_dataset(0,20)
```

```
def preprocess_dataset(dataset, test_size=0.1, val_size=0.2):
  x = np.array([normalize_vector(i) for i in dataset['x']], dtype=np.float)
  x = np.reshape(x, (x.shape[0], x.shape[1], 1))
  y = np.array(dataset['y'], dtype=np.uint8)

  ohe = OneHotEncoder()
  y = ohe.fit_transform(y.reshape((-1,1))).toarray()
  
  x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=test_size, random_state=42)
  x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=val_size, random_state=42)

  print(x_train.shape, y_train.shape)
  print(x_val.shape, y_val.shape)
  print(x_test.shape, y_test.shape)

  return x_train, x_val, x_test, y_train, y_val, y_test
```

```
x_train, x_val, x_test, y_train, y_val, y_test = preprocess_dataset(dataset, 0.05, 0.21)
```

###MODEL TEMPORAL CNN

```
def build_model(time_steps, features, latent_dim=128, dilation_rate=8):
  i = Input(shape=(time_steps, features))
  x = i
  x = Conv1D(16, 3, strides=2, padding='same')(x)
  x = LeakyReLU(alpha=0.2)(x)
  x = BatchNormalization()(x)
  
  x = Conv1D(32, 3, strides=2, padding='same')(x)
  x = LeakyReLU(alpha=0.2)(x)
  x = BatchNormalization()(x)
  
  x = Conv1D(64, 3, strides=2, activation='relu', padding='same')(x)
  x = LeakyReLU(alpha=0.2)(x)
  x = BatchNormalization()(x)
  
  volume_size = K.int_shape(x)

  latent = TCN(nb_filters=latent_dim,
        kernel_size=4,
        dilations=[2**i for i in range(dilation_rate)])(x)
  
  encoder = Model(i, latent, name="encoder")

  latent_inputs = Input(shape=(latent_dim,))
  x = Dense(np.prod(volume_size[1:]))(latent_inputs)
  x = Reshape((volume_size[1], volume_size[2]))(x)
  
  x = Conv1D(64, 3, activation='relu', padding='same')(x)
  x = UpSampling1D(size=2)(x)
  x = LeakyReLU(alpha=0.2)(x)
  x = BatchNormalization()(x)  
  
  x = Conv1D(32, 3, activation='relu', padding='same')(x)
  x = UpSampling1D(size=2)(x)
  x = LeakyReLU(alpha=0.2)(x)
  x = BatchNormalization()(x)

  x = Conv1D(16, 3, activation='relu', padding='same')(x)
  x = UpSampling1D(size=2)(x)
  x = LeakyReLU(alpha=0.2)(x)
  x = BatchNormalization()(x)

  o = Conv1D(1, 3, activation='sigmoid', padding='same')(x)

  decoder = Model(latent_inputs, o, name="decoder")

  autoencoder = Model(i, decoder(encoder(i)), name="autoencoder")
  return (encoder, decoder, autoencoder)
```

```
(encoder, decoder, autoencoder) = build_model(3000, 1, 512, 7)
autoencoder.summary()
```

```
INIT_LR = 0.0005
EPOCHS = 15
BATCH_SIZE = 16
STEPS_PER_EPOCH = x_train.shape[0] // BATCH_SIZE
VALIDATION_STEPS = x_val.shape[0] // BATCH_SIZE

def poly_decay(epoch):
  maxEpochs = EPOCHS
  baseLR = INIT_LR
  power = 1.0

  alpha = baseLR * (1 - (epoch / float(maxEpochs))) ** power
  return alpha

callbacks = [LearningRateScheduler(poly_decay)]
```

```
def acc(y_true, y_pred):
  return K.mean(K.equal(K.round(y_true), K.round(y_pred)))

opt = Adam(lr=INIT_LR, amsgrad=True)
autoencoder.compile(loss='mse', optimizer=opt, metrics=[acc])
```

```
r = autoencoder.fit(
  x_train, x_train,
  validation_data=(x_val, x_val),
  epochs=EPOCHS, batch_size=BATCH_SIZE,
  callbacks=callbacks
)
```
